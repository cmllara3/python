# Modelos de regresión para el precio de una casa

## Importación de librerias

<b>Paso 0.</b> Importación de librerías

```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## ETL

<b>Paso 1.</b> Cargue de los archivos

```python
#Import data train_1
train_1 = pd.read_csv("../01. Flat Files/train_1.csv",delimiter=",")
print("Archivo train_1 cargado: ",train_1.shape)
print(train_1.dtypes)
print("")

#Import data train_2
train_2 = pd.read_csv("../01. Flat Files/train_2.csv",delimiter=",")
print("Archivo train_2 cargado: ",train_2.shape)
print(train_2.dtypes)
print("")

#Import data predict_1
predict_1 = pd.read_csv("../01. Flat Files/predict_1.csv",delimiter=",")
print("Archivo predict_1 cargado: ",predict_1.shape)
print(predict_1.dtypes)
print("")

#Import data predict_2
predict_2 = pd.read_csv("../01. Flat Files/predict_2.csv",delimiter=",")
print("Archivo predict_2 cargado: ",predict_2.shape)
print(predict_2.dtypes)
```

<b>Paso 2.</b> Union de los DataFrame

```python
#Se unen los dos data frame de entrenamiento por columnas haciendo join por la columna ID
dataTrain = pd.concat([train_1,train_2],axis=1,join="inner")

#Se unen los dos data frame de prediccion por columnas haciendo join por la columna ID
dataPredict = pd.concat([predict_1,predict_2],axis=1,join="inner")
```

<b>Paso 3.</b> Validacion de informacion faltante

```python
#Validacion campos faltantes datos de entrenamiento
print("dataTrain")
print(dataTrain.isnull().any())
print("")

#Validacion campos faltantes datos de prediccion
print("dataPredict")
print(dataPredict.isnull().any())
```

<b>Paso 4.</b> Histograma de frecuencia para ver el comportamiento de cada una de las variables. Se excluye las variables: a) chas --> Variable binaria, b) id --> Varaible label y c) medv --> Variable a predecir

```python
#Particion del grafico en 5 fila y 3 columnas
figure,axes = plt.subplots(nrows=4,ncols=3,figsize=(20,20))
figure.tight_layout(pad=3.0)

#Histograma de frecuencias zn
ax = axes[0][0]
ax.title.set_text("Var: zn")
ax.hist(dataTrain["zn"],bins=10)
ax.grid(True)

#Histograma de frecuencias rm
ax = axes[0][1]
ax.title.set_text("Var: rm")
ax.hist(dataTrain["rm"],bins=10)
ax.grid(True)

#Histograma de frecuencias dis
ax = axes[0][2]
ax.title.set_text("Var: dis")
ax.hist(dataTrain["dis"],bins=10)
ax.grid(True)

#Histograma de frecuencias rad
ax = axes[1][0]
ax.title.set_text("Var: rad")
ax.hist(dataTrain["rad"],bins=10)
ax.grid(True)

#Histograma de frecuencias tax
ax = axes[1][1]
ax.title.set_text("Var: tax")
ax.hist(dataTrain["tax"],bins=10)
ax.grid(True)

#Histograma de frecuencias crim
ax = axes[1][2]
ax.title.set_text("Var: crim")
ax.hist(dataTrain["crim"],bins=10)
ax.grid(True)

#Histograma de frecuencias indus
ax = axes[2][0]
ax.title.set_text("Var: indus")
ax.hist(dataTrain["indus"],bins=10)
ax.grid(True)

#Histograma de frecuencias nox
ax = axes[2][1]
ax.title.set_text("Var: nox")
ax.hist(dataTrain["nox"],bins=10)
ax.grid(True)

#Histograma de frecuencias age
ax = axes[2][2]
ax.title.set_text("Var: age")
ax.hist(dataTrain["age"],bins=10)
ax.grid(True)

#Histograma de frecuencias ptratio
ax = axes[3][0]
ax.title.set_text("Var: ptratio")
ax.hist(dataTrain["ptratio"],bins=10)
ax.grid(True)

#Histograma de frecuencias black
ax = axes[3][1]
ax.title.set_text("Var: black")
ax.hist(dataTrain["black"],bins=10)
ax.grid(True)

#Histograma de frecuencias lstat
ax = axes[3][2]
ax.title.set_text("Var: lstat")
ax.hist(dataTrain["age"],bins=10)
ax.grid(True)

plt.show()
```

## Modelos de regresión supervisados

<b>Paso 5.</b> Se realiza la particion de los datos para obtener los datos de testeo y entrenamiento

```python
#Separacion de los datos y la variable a predecir
x = dataTrain.drop(columns={"ID","medv"})
y = dataTrain["medv"]

#Particion de los datos para testeo y entrenamiento
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(x,y,test_size=0.3,random_state=101)
```

<b>Paso 6.</b> Aplicación de modelo de regresión. Dado que la cantidad de los datos es relativamente pequeña se realizaran pruebas sobre estos datos con diferentes modelos para determinar cual se ajusta mejor.

<b>Modelo de regresion lineal</b>


```python
#Se importa la ejecucion secuencial 
from sklearn.pipeline import Pipeline
#Funcion para la normalizacion de variables
from sklearn.preprocessing import Normalizer
#Importacion del modelo de regresion lineal
from sklearn.linear_model import LinearRegression

#Se inicializa el modelo normalizando primero y aplicando despues el modelo de regresion lineal
modeloRLN = Pipeline([("normalizacion",Normalizer()),("modelo",LinearRegression())])

#Se aplica el modelo de regresion lineal sobre los datos de entrenamiento
modeloRLN.fit(xTrain, yTrain)
```

```python
#Predicciones para regresion lineal
predictionsRLN = modeloRLN.predict(xTest)
```

<b>Modelo de arboles de regresion</b>

```python
#Se importa la ejecucion secuencial 
from sklearn.pipeline import Pipeline
#Funcion para la normalizacion de variables
from sklearn.preprocessing import Normalizer
#Importacion del modelo de arboles de regresion
from sklearn.tree import DecisionTreeRegressor

#Se inicializa el modelo normalizando primero y aplicando despues el modelo de arbol de regresion
modeloAR = Pipeline([("normalizacion",Normalizer()),("modelo",DecisionTreeRegressor())])

#Se aplica el modelo de arbol de regresion sobre los datos de entrenamiento
modeloAR.fit(xTrain, yTrain)
```

```python
#Predicciones para arbol de regresion
predictionsAR = modeloAR.predict(xTest)
```

<b>Modelo de random forest</b>

```python
#Se importa la ejecucion secuencial 
from sklearn.pipeline import Pipeline
#Funcion para la normalizacion de variables
from sklearn.preprocessing import Normalizer
#Importacion del modelo de random forest
from sklearn.ensemble import RandomForestRegressor

#Se inicializa el modelo normalizando primero y aplicando despues el modelo de random forest
modeloRF = Pipeline([("normalizacion",Normalizer()),("modelo",RandomForestRegressor())])

#Se aplica el modelo de random forest sobre los datos de entrenamiento
modeloRF.fit(xTrain, yTrain)
```

```python
#Predicciones para random forest
predictionsRF = modeloRF.predict(xTest)
```

<b>Modelos de SVM</b>

```python
#Se importa la ejecucion secuencial 
from sklearn.pipeline import Pipeline
#Funcion para la normalizacion de variables
from sklearn.preprocessing import Normalizer
#Importacion del modelo de SVM para regresion
from sklearn.svm import LinearSVR

#Se inicializa el modelo normalizando primero y aplicando despues el modelo de SVM
modeloSVM = Pipeline([("normalizacion",Normalizer()),("modelo",LinearSVR())])

#Se aplica el modelo de SVM sobre los datos de entrenamiento
modeloSVM.fit(xTrain, yTrain)
```

```python
#Predicciones para SVM
predictionsSVM = modeloSVM.predict(xTest)
```

<b>Comparacion de modelos</b>

```python
#Se crea el DF donde se almacenara las varaibles de comparación. 
#maxError --> Estadistica de la diferencia del mayor error
#MAE --> El promedio de los errores
#MSE --> El promedio de los errores al cuadradro
#r2 --> Explica que tanto porcentaje de la varianza de la totalidad del modelo es explicada por las varaibles en relacion
comparacion = pd.DataFrame(columns=["modelo","maxError","MAE","MSE","r2"],index=range(4))

#Se agrega el nombre de cada uno de los modelos
comparacion.iloc[0][0] = "Regresion Lineal"
comparacion.iloc[1][0] = "Arbol de Regresion"
comparacion.iloc[2][0] = "Random Forest"
comparacion.iloc[3][0] = "SVM"

#Se importa la metrica de maxError
from sklearn.metrics import max_error

#Se actualiza el campo de maxError
comparacion.iloc[0][1] = round(max_error(yTest,predictionsRLN),3)
comparacion.iloc[1][1] = round(max_error(yTest,predictionsAR),3)
comparacion.iloc[2][1] = round(max_error(yTest,predictionsRF),3)
comparacion.iloc[3][1] = round(max_error(yTest,predictionsSVM),3)

#Se importa la metrica de MAE
from sklearn.metrics import mean_absolute_error

#Se actualiza el campo de maxError
comparacion.iloc[0][2] = round(mean_absolute_error(yTest,predictionsRLN),3)
comparacion.iloc[1][2] = round(mean_absolute_error(yTest,predictionsAR),3)
comparacion.iloc[2][2] = round(mean_absolute_error(yTest,predictionsRF),3)
comparacion.iloc[3][2] = round(mean_absolute_error(yTest,predictionsSVM),3)

#Se importa la metrica de MSE
from sklearn.metrics import mean_squared_error

#Se actualiza el campo de MSE
comparacion.iloc[0][3] = round(mean_squared_error(yTest,predictionsRLN),3)
comparacion.iloc[1][3] = round(mean_squared_error(yTest,predictionsAR),3)
comparacion.iloc[2][3] = round(mean_squared_error(yTest,predictionsRF),3)
comparacion.iloc[3][3] = round(mean_squared_error(yTest,predictionsSVM),3)

#Se importa la metrica de R2
from sklearn.metrics import r2_score

#Se actualiza el campo de R2
comparacion.iloc[0][4] = round(r2_score(yTest,predictionsRLN),3)
comparacion.iloc[1][4] = round(r2_score(yTest,predictionsAR),3)
comparacion.iloc[2][4] = round(r2_score(yTest,predictionsRF),3)
comparacion.iloc[3][4] = round(r2_score(yTest,predictionsSVM),3)

comparacion
```

A partir de los resultados obtenidos se puede afirmar que el modelo que mejor se ajusta a nuestros datos teniendo en cuenta mayormente el r2 es el Random Forest que se utilizara para predecir el precio de los inmuebles

## Aplicación modelo de predicción

<b>Paso 7.</b> Se realiza la prediccion sobre los datos entregados inicialmente

```python
#Separacion de los datos y la variable a predecir
xPredict = dataPredict.drop(columns={"ID"})

#Predicciones para random forest
yPredict = modeloRF.predict(xPredict)

#Se crea el data frame con el Id de la casa y el precio
exportar = pd.concat([dataPredict.iloc[:,0],pd.DataFrame(data=yPredict,columns={"medv"})],axis=1)

#Se exporta el DF en CSV
exportar.to_csv("../01. Flat Files/resultado_pred.csv",index=False)
```
