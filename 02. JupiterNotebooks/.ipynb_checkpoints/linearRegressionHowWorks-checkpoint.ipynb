<b>Paso 5. </b> Definimos la función de Regresión Lineal. <br>
<i> Tres parámetros: x(altura), w(peso), b(bias). Por defecto w y b tienen un valor de 0

```python
def regresionLinealManual (x,w=0,b=0):
    return b+(w*x)
```

<b>Paso 7.</b> Creamos un array yhat resultado de aplicar sobre x la función definida junto los parámetros w y b. Con w=0 y b=0

```python
weigth1 = regresionLinealManual(heightPrueba)
```

<b>Paso 8.</b> Mostramos la gráfica resultante con w=0 y b=0

```python
plt.plot(heightPrueba,weigth1,'r',linewidth=3)

plt.show()
```

```python
weigth2 = regresionLinealManual(heightPrueba,w=2,b=50)

plt.xlabel('height', fontsize=14)
plt.ylabel('weight', fontsize=14)
plt.title('scatter plot - height vs weight',fontsize=20)
plt.scatter(weight,height)

plt.plot(heightPrueba,weigth1,'r',linewidth=3,label = "y=0x+b")
plt.plot(heightPrueba,weigth2,'b',linewidth=3,label = "y=2x+50")

plt.legend(loc='best')

plt.show()
```

### Cost Function

<b>Paso 0.</b> Definimos la función que calcula el error (Mean Squared Error)

```python
def mse (yReal,yPredict):
    err2 = (yReal-yPredict)**2
    return err2.mean()
```

<b>Paso 1.</b> Especificamos qué es X (alturas leídas del fichero) y qué es y_true (pesos leídos del fichero)

```python
x = data["Height"]
yReal = data["Weight"]
```

<b>Paso 2.</b> Calculamos predicción con w=0 y b=0

```python
yPredict = regresionLinealManual(x)
```

<b>Paso 3.</b> Calculamos el error resultante para y_pred

```python
print(mse(yReal,yPredict))
```

<b>Paso 4.</b> Prueba tu!!! Prueba 3 valores distintos de w y b. Muestra en un gráfico los resultados y la función de coste

```python
yPredict = regresionLinealManual(x,w=2,b=50)
print(mse(yReal,yPredict))
print("")

yPredict = regresionLinealManual(x,w=1,b=57)
print(mse(yReal,yPredict))
print("")

yPredict = regresionLinealManual(x,w=3,b=45)
print(mse(yReal,yPredict))
```

<b>Paso 5.</b> Vamos a mostrar cómo automatizar un proceso de cálculos sobre distintos w y b, con su respectiva función de coste

```python
figure,axes = plt.subplots(nrows=1, ncols=2)
figure.tight_layout(pad=3.0)
figure.figsize=(10,5)

ax1 = axes[0]

ax1.scatter(x,yReal)
ax1.title.set_text("height vs weight")
ax1.set_xlabel('height')
ax1.set_ylabel('weight')

arrayBValues = np.array([-100,-50,0,50,100,150])
arrayErr = []

for b in arrayBValues:
    yPredict = regresionLinealManual(x,w=2,b=b)
    mseValue = mse(yReal,yPredict)
    arrayErr.append(mseValue)
    lab = "y=2x+"+str(b)
    ax1.plot(x,yPredict,label=lab)
     
ax1.legend(loc='best')

ax2 = axes[1]
ax2.title.set_text("Costo por b")
ax2.set_xlabel('b')
ax2.set_ylabel('MSE')
ax2.plot(arrayBValues,arrayErr,'o-')

plt.show()
```
